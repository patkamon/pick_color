```python
from PIL import Image
from collections import defaultdict
import operator
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_rgb_colors(image_path):
    # Open the image using the Image module
    image = Image.open(image_path)

    # Count the occurrences of each RGB color
    color_counts = defaultdict(int)
    for x in range(image.width):
        for y in range(image.height):
            color = image.getpixel((x, y))
            color_counts[color] += 1

    # Sort the color counts in descending order
    sorted_colors = sorted(color_counts.items(), key=operator.itemgetter(1), reverse=True)

    # Print the RGB colors in order from most to least frequent
    for color, count in sorted_colors[:10]:
        print(f"RGB color: {color}, count: {count}")
    return sorted_colors[1:10]
```

```python
def plot_colors(list):
    i = 1
    fig = plt.figure(figsize=(10,4))
    for c, _ in list:
        plt.subplot(2,5,i)
        i+=1
        plt.imshow([[c]])
        plt.xlabel(str(_) + " " + str(c))
    plt.tight_layout()
    plt.show()
```

```python
def replace_color(image_name, origin_c, replace_c):
    orig_color = origin_c[0:3]
    replacement_color =  replace_c
    img = Image.open(image_name).convert('RGB')
    data = np.array(img)
    data[(data == orig_color).all(axis = -1)] = replacement_color
    img2 = Image.fromarray(data, mode='RGB')
    display(img2)
```

```python
# Example usage
colors = get_rgb_colors("balrog.png")
colors
```

```python
plot_colors(colors)
```

```python
for c,_ in colors[:4]:
    print(c,_)
    replace_color("balrog.png", c, (0,0,255)) #blue
```
